```python
import gradio as gr
from transformers import pipeline
```

```python
translation_pipeline = pipeline("translation_en_to_de")
result = translation_pipeline("I like mathematics and computers")
```

```python
result[0]["translation_text"]
```

```python
def translate_gradio(input):
  result=translation_pipeline(input)
  return result[0]["translation_text"]
```

```python
translate_interface=gr.Interface(fn=translate_gradio,inputs="text",outputs="text")
btn2 = gr.Button("Translate")
```

```python
translate_interface.launch()
```
